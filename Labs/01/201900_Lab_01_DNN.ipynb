```python
import numpy as np

# Define the ODE function dy/dx = f(x, y)
def f(x, y):
    return -2 * x * y

# Define the initial condition
y0 = [1.0]

# Define the range of x values
x_span = (0, 2)

# Solve the ODE using the solve_ivp function
solution = solve_ivp(f, x_span, y0, t_eval=np.linspace(x_span[0], x_span[1], 100))

# Extract the solution values
x_values = solution.t
y_values = solution.y[0]
```

```python
# Create a time vector
time = np.linspace(0, 1, 1000)

# Define parameters for individual sine waves
frequencies = [3, 5, 7]  # Frequencies of the sine waves (3 Hz, 5 Hz, 7 Hz)
amplitudes = [1.0, 0.5, 0.3]  # Amplitudes of the sine waves

# Create a compound signal by summing individual sine waves
compound_signal = sum(amplitude * np.sin(2 * np.pi * frequency * time) for amplitude, frequency in zip(amplitudes, frequencies))

# Apply FFT to the compound signal
signal_fft = fft(compound_signal)

# Apply IFFT to the FFT result to recover the original signal
recovered_signal = ifft(signal_fft)

# Print the original compound signal and the recovered signal
print("Original Compound Signal:")
print(compound_signal[:10])  # Print the first 10 samples of the original compound signal
print("\nRecovered Signal:")
print(recovered_signal[:10].real)  # Print the first 10 samples of the recovered signal
```

```python
import numpy as np

# Function to initialize neural network parameters randomly
def initialize_parameters(input_size, hidden_size, output_size):
    # Seed for reproducibility
    np.random.seed(0)
    # Initialize weights and biases randomly
    W1 = np.random.randn(hidden_size, input_size)
    b1 = np.random.randn(hidden_size, 1)
    W2 = np.random.randn(output_size, hidden_size)
    b2 = np.random.randn(output_size, 1)
    return W1, b1, W2, b2

# Activation function (identity function in this case)
def activation(x):
    return x

# Forward propagation through the neural network
def forward_propagation(X, W1, b1, W2, b2):
    # Calculate first layer activations
    Z1 = np.dot(W1, X) + b1
    A1 = activation(Z1)
    # Calculate output layer activations
    Z2 = np.dot(W2, A1) + b2
    return Z2

# Loss function (Mean Squared Error)
def loss_function(predictions, targets):
    return np.mean((predictions - targets) ** 2)

# Function to train the neural network
def train_neural_network(X_train, y_train, W1, b1, W2, b2, learning_rate, num_iterations):
    for i in range(num_iterations):
        # Forward propagation
        predictions = forward_propagation(X_train, W1, b1, W2, b2)
        # Compute loss
        loss = loss_function(predictions, y_train)
        # Backpropagation
        dZ2 = predictions - y_train
        dW2 = np.dot(dZ2, forward_propagation(X_train, W1, b1, W2, b2).T) / X_train.shape[1]
        db2 = np.mean(dZ2, axis=1, keepdims=True)
        dZ1 = np.dot(W2.T, dZ2) * activation(forward_propagation(X_train, W1, b1, W2, b2))
        dW1 = np.dot(dZ1, X_train.T) / X_train.shape[1]
        db1 = np.mean(dZ1, axis=1, keepdims=True)
        # Update weights and biases
        W1 -= learning_rate * dW1
        b1 -= learning_rate * db1
        W2 -= learning_rate * dW2
        b2 -= learning_rate * db2
        # Print the loss every 10 iterations
        if i % 10 == 0:
            print(f"Iteration {i}, Loss: {loss}")
    return W1, b1, W2, b2

# Function to make predictions using the trained neural network
def predict(X_test, W1, b1, W2, b2):
    return forward_propagation(X_test, W1, b1, W2, b2)

# Usage
input_size = 1
hidden_size = 3
output_size = 1
learning_rate = 0.0001
num_iterations = 100

# Generate random training data
X_train = np.random.rand(1, 1000) * 100
y_train = 2.5 * X_train + 10 + np.random.randn(1, 1000) * 10

# Initialize parameters
W1, b1, W2, b2 = initialize_parameters(input_size, hidden_size, output_size)

# Train the neural network
W1, b1, W2, b2 = train_neural_network(X_train, y_train, W1, b1, W2, b2, learning_rate, num_iterations)

# Test the trained model on new data
X_test = np.array([[1500]])  # New data point
predicted_price = predict(X_test, W1, b1, W2, b2)
print("Predicted price:", predicted_price[0][0])
```

```python
import numpy as np

# Function to initialize neural network parameters randomly
def initialize_parameters(input_size, hidden_size, output_size):
    # Seed for reproducibility
    np.random.seed(0)
    # Initialize weights and biases randomly
    W1 = np.random.randn(hidden_size, input_size)
    b1 = np.random.randn(hidden_size, 1)
    W2 = np.random.randn(output_size, hidden_size)
    b2 = np.random.randn(output_size, 1)
    return W1, b1, W2, b2

# ReLU activation function
def activation(x):
    return np.maximum(0, x)

# Forward propagation through the neural network
def forward_propagation(X, W1, b1, W2, b2):
    # Calculate activations for the first hidden layer
    Z1 = np.dot(W1, X) + b1
    A1 = activation(Z1)
    # Calculate output layer activations
    Z2 = np.dot(W2, A1) + b2
    return Z2

# Loss function (Mean Squared Error)
def loss_function(predictions, targets):
    return np.mean((predictions - targets) ** 2)

# Function to train the neural network
def train_neural_network(X_train, y_train, W1, b1, W2, b2, learning_rate, num_iterations):
    for i in range(num_iterations):
        # Forward propagation
        predictions = forward_propagation(X_train, W1, b1, W2, b2)
        # Compute loss
        loss = loss_function(predictions, y_train)

        # Backpropagation
        dZ2 = predictions - y_train
        dW2 = np.dot(dZ2, forward_propagation(X_train, W1, b1, W2, b2).T) / X_train.shape[1]
        db2 = np.mean(dZ2, axis=1, keepdims=True)
        dZ1 = np.dot(W2.T, dZ2) * (forward_propagation(X_train, W1, b1, W2, b2) > 0)  # Derivative of ReLU
        dW1 = np.dot(dZ1, X_train.T) / X_train.shape[1]
        db1 = np.mean(dZ1, axis=1, keepdims=True)

        # Update weights and biases
        W1 -= learning_rate * dW1
        b1 -= learning_rate * db1
        W2 -= learning_rate * dW2
        b2 -= learning_rate * db2

        # Print the loss every 100 iterations
        if i % 100 == 0:
            print(f"Iteration {i}, Loss: {loss}")

    return W1, b1, W2, b2

# Function to make predictions using the trained neural network
def predict(X_test, W1, b1, W2, b2):
    return forward_propagation(X_test, W1, b1, W2, b2)

# Usage
input_size = 1
hidden_size = 3
output_size = 1
learning_rate = 0.001
num_iterations = 1000

# Generate random training data
X_train = np.random.rand(1, 1000) * 100
y_train = 2.5 * X_train + 10 + np.random.randn(1, 1000) * 10

# Initialize parameters
W1, b1, W2, b2 = initialize_parameters(input_size, hidden_size, output_size)

# Train the neural network
W1, b1, W2, b2 = train_neural_network(X_train, y_train, W1, b1, W2, b2, learning_rate, num_iterations)

# Test the trained model on new data
X_test = np.array([[1500]])  # New data point
predicted_price = predict(X_test, W1, b1, W2, b2)
print("Predicted price:", predicted_price[0][0])
```

```python
import numpy as np

# Function to initialize neural network parameters using Xavier initialization
def initialize_parameters(input_size, hidden_sizes, output_size):
    np.random.seed(0)
    parameters = {}
    layer_sizes = [input_size] + hidden_sizes + [output_size]
    for i in range(1, len(layer_sizes)):
        parameters[f'W{i}'] = np.random.randn(layer_sizes[i], layer_sizes[i-1]) * np.sqrt(2 / layer_sizes[i-1])
        parameters[f'b{i}'] = np.zeros((layer_sizes[i], 1))
    return parameters

# ReLU activation function and its derivative
def activation(x):
    return np.maximum(0, x)

def activation_derivative(x):
    return np.where(x > 0, 1, 0)

# Mean Squared Error loss function
def loss_function(predictions, targets):
    return np.mean((predictions - targets) ** 2)

# Forward propagation through the neural network
def forward_propagation(X, parameters):
    cache = {'A0': X}
    L = len(parameters) // 2
    for l in range(1, L + 1):
        cache[f'Z{l}'] = np.dot(parameters[f'W{l}'], cache[f'A{l-1}']) + parameters[f'b{l}']
        if l == 1:
            cache[f'A{l}'] = activation(cache[f'Z{l}'])
        else:
            cache[f'A{l}'] = cache[f'Z{l}']  # Linear activation for the second hidden layer
    return cache[f'A{L}'], cache

# Backpropagation
def backward_propagation(parameters, cache, X, Y):
    grads = {}
    L = len(parameters) // 2
    m = X.shape[1]
    dAL = 2 * (cache[f'A{L}'] - Y) / m
    for l in range(L, 0, -1):
        dZL = dAL * activation_derivative(cache[f'Z{l}'])
        grads[f'dW{l}'] = np.dot(dZL, cache[f'A{l-1}'].T)
        grads[f'db{l}'] = np.sum(dZL, axis=1, keepdims=True)
        if l > 1:
            dAL = np.dot(parameters[f'W{l}'].T, dZL)
    return grads

# Update parameters using gradient descent
def update_parameters(parameters, grads, learning_rate):
    L = len(parameters) // 2
    for l in range(1, L + 1):
        parameters[f'W{l}'] -= learning_rate * grads[f'dW{l}']
        parameters[f'b{l}'] -= learning_rate * grads[f'db{l}']
    return parameters

# Function to train the neural network
def train_neural_network(X_train, y_train, parameters, learning_rate, num_iterations):
    for i in range(num_iterations):
        # Forward propagation
        predictions, cache = forward_propagation(X_train, parameters)
        # Compute loss
        loss = loss_function(predictions, y_train)
        # Backpropagation
        grads = backward_propagation(parameters, cache, X_train, y_train)
        # Update parameters
        parameters = update_parameters(parameters, grads, learning_rate)
        # Print the loss every 100 iterations
        if i % 100 == 0:
            print(f"Iteration {i}, Loss: {loss}")
    return parameters

# Function to make predictions using the trained neural network
def predict(X_test, parameters):
    predictions, _ = forward_propagation(X_test, parameters)
    return predictions

# Usage
input_size = 1
hidden_sizes = [5, 3]  # Change hidden layer sizes here
output_size = 1
learning_rate = 0.001  # Decreased learning rate
num_iterations = 1000

# Generate random training data
X_train = np.random.rand(1, 1000) * 100
y_train = 2.5 * X_train + 10 + np.random.randn(1, 1000) * 10

# Initialize parameters using Xavier initialization
parameters = initialize_parameters(input_size, hidden_sizes, output_size)

# Train the neural network
parameters = train_neural_network(X_train, y_train, parameters, learning_rate, num_iterations)

# Test the trained model on new data
X_test = np.array([[1500]])  # New data point
predicted_price = predict(X_test, parameters)
print("Predicted price:", predicted_price[0][0])
```


 1. **Normalization Function (`normalize_data`)**: Normalizes the input data to have a mean of zero and standard deviation of one along each feature.

 2. **Parameter Initialization Function (`initialize_parameters`)**: Initializes the weights and biases of the neural network using Xavier initialization, ensuring stable training.

 3. **Activation Functions (`activation` and `activation_derivative`)**: Implements the Rectified Linear Unit (ReLU) activation function and its derivative.

 4. **Loss Function (`loss_function`)**: Computes the Mean Squared Error (MSE) loss with L2 regularization to prevent overfitting.

 5. **Forward Propagation (`forward_propagation`)**: Performs forward propagation through the neural network, computing activations for each layer.

 6. **Backward Propagation (`backward_propagation`)**: Computes the gradients of the parameters using backpropagation with L2 regularization.

 7. **Parameter Update Function (`update_parameters`)**: Updates the parameters (weights and biases) of the neural network using gradient descent.

 8. **Training Function (`train_neural_network`)**: Trains the neural network using forward and backward propagation for a specified number of iterations.

 9. **Prediction Function (`predict`)**: Makes predictions using the trained neural network on new data.


```python
import numpy as np

# Function to normalize input data
def normalize_data(X):
    mean = np.mean(X, axis=1, keepdims=True)
    std = np.std(X, axis=1, keepdims=True)
    X_normalized = (X - mean) / std
    return X_normalized

# Function to initialize neural network parameters using Xavier initialization
def initialize_parameters(input_size, hidden_sizes, output_size):
    np.random.seed(0)
    parameters = {}
    layer_sizes = [input_size] + hidden_sizes + [output_size]
    for i in range(1, len(layer_sizes)):
        parameters[f'W{i}'] = np.random.randn(layer_sizes[i], layer_sizes[i-1]) * np.sqrt(2 / layer_sizes[i-1])
        parameters[f'b{i}'] = np.zeros((layer_sizes[i], 1))
    return parameters

# ReLU activation function and its derivative
def activation(x):
    return np.maximum(0, x)

def activation_derivative(x):
    return np.where(x > 0, 1, 0)

# Loss function (Mean Squared Error) with L2 regularization
def loss_function(predictions, targets, parameters, lambd):
    m = predictions.shape[1]
    mse_loss = np.mean((predictions - targets) ** 2)
    l2_regularization = 0.5 * lambd * sum(np.sum(np.square(parameters[f'W{i}'])) for i in range(1, len(parameters) // 2 + 1))
    total_loss = mse_loss + l2_regularization
    return total_loss

# Forward propagation through the neural network
def forward_propagation(X, parameters):
    cache = {'A0': X}
    L = len(parameters) // 2
    for l in range(1, L + 1):
        cache[f'Z{l}'] = np.dot(parameters[f'W{l}'], cache[f'A{l-1}']) + parameters[f'b{l}']
        if l == 1:
            cache[f'A{l}'] = activation(cache[f'Z{l}'])
        else:
            cache[f'A{l}'] = cache[f'Z{l}']  # Linear activation for the second hidden layer
    return cache[f'A{L}'], cache

# Backpropagation with L2 regularization
def backward_propagation(parameters, cache, X, Y, lambd):
    grads = {}
    L = len(parameters) // 2
    m = X.shape[1]
    dAL = 2 * (cache[f'A{L}'] - Y) / m
    for l in range(L, 0, -1):
        dZL = dAL * activation_derivative(cache[f'Z{l}'])
        grads[f'dW{l}'] = np.dot(dZL, cache[f'A{l-1}'].T) + lambd * parameters[f'W{l}']
        grads[f'db{l}'] = np.sum(dZL, axis=1, keepdims=True)
        if l > 1:
            dAL = np.dot(parameters[f'W{l}'].T, dZL)
    return grads

# Update parameters
def update_parameters(parameters, grads, learning_rate):
    L = len(parameters) // 2
    for l in range(1, L + 1):
        parameters[f'W{l}'] -= learning_rate * grads[f'dW{l}']
        parameters[f'b{l}'] -= learning_rate * grads[f'db{l}']
    return parameters

# Function to train the neural network
def train_neural_network(X_train, y_train, parameters, learning_rate, num_iterations, lambd):
    loss_history = []
    for i in range(num_iterations):
        # Forward propagation
        predictions, cache = forward_propagation(X_train, parameters)
        # Compute loss
        loss = loss_function(predictions, y_train, parameters, lambd)
        loss_history.append(loss)
        # Backpropagation
        grads = backward_propagation(parameters, cache, X_train, y_train, lambd)
        # Update parameters
        parameters = update_parameters(parameters, grads, learning_rate)
        # Print the loss every 100 iterations
        if i % 100 == 0:
            print(f"Iteration {i}, Loss: {loss}")
    return parameters, loss_history

# Function to make predictions using the trained neural network
def predict(X_test, parameters):
    predictions, _ = forward_propagation(X_test, parameters)
    return predictions

# Usage
input_size = 1
hidden_sizes = [5, 3]  # Change hidden layer sizes here
output_size = 1
learning_rate = 0.001
num_iterations = 1000
lambd = 0.01  # Regularization parameter

# Generate random training data
X_train = np.random.rand(1, 1000) * 100
y_train = 2.5 * X_train + 10 + np.random.randn(1, 1000) * 10

# Normalize input data
X_train_normalized = normalize_data(X_train)

# Initialize parameters using Xavier initialization
parameters = initialize_parameters(input_size, hidden_sizes, output_size)

# Train the neural network
parameters, loss_history = train_neural_network(X_train_normalized, y_train, parameters, learning_rate, num_iterations, lambd)

# Test the trained model on new data
X_test = np.array([[1500]])  # New data point
predicted_price = predict(X_test, parameters)
print("Predicted price:", predicted_price[0][0])
```
